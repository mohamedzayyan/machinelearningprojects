```python
train_transf = transforms.Compose([transforms.RandomHorizontalFlip(p=0.5),
                             transforms.RandomRotation(degrees=(0,360)),
                             transforms.RandomVerticalFlip(p=0.5),
                             transforms.ToTensor(),
                             transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                                  std=[0.229, 0.224, 0.225])])
```

```python
test_transf = transforms.Compose([transforms.ToTensor(),
                             transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                                  std=[0.229, 0.224, 0.225])])
```

```python
device = torch.device('cpu')
```

```python
cifar_trainset = datasets.CIFAR10(root='./data', train=True, download=True, transform=train_transf)
```

```python
cifar_testset = datasets.CIFAR10(root='./data', train=False, download=True, transform=test_transf)
```

```python
size = len(cifar_trainset)
indices = list(range(size))
split = int(np.floor(0.2 * size))
if True:
    np.random.seed(42)
    np.random.shuffle(indices)
train_indices, val_indices = indices[split:], indices[:split]
test_indices = list(range(len(cifar_testset)))

train_sampler = SubsetRandomSampler(train_indices)
val_sampler   = SubsetRandomSampler(val_indices)
test_sampler = SubsetRandomSampler(test_indices)


train_dl = DataLoader(cifar_trainset, batch_size=32, num_workers=0,
                                                 sampler=train_sampler)
val_dl = DataLoader(cifar_trainset, batch_size=32, num_workers=0,
                                                 sampler=val_sampler)
unseen_dl = DataLoader(cifar_testset,batch_size=32, num_workers=0,
                                                 sampler=test_sampler)
print("Total: {} Train_dl: {} Validation_dl: {} Test_dl: {}".format(size, len(train_dl),
                                                                    len(val_dl),
                                                                   len(unseen_dl)))
```

```python
def train_phase():
    model.train()
    losses = []
    epoch_start = time.time()
    for i, batch in enumerate(train_dl):
        inputs, labels = batch[0], batch[1].long()
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()             # 1. Zero the parameter gradients
        outputs = model(inputs)           # 2. Run the model

        loss = criterion(outputs, labels) # 3. Calculate loss
        #loss = criterion(outputs, torch.max(labels, 1)[1])
        losses.append(loss.item())
        #loss = torch.sqrt(loss)           #    -> RMSE loss
        loss.backward()                   # 4. Backward propagate the loss
        optimizer.step()                  # 5. Optimize the network

        #print("--> Batch {}/{} Loss: {}".format(i+1, len(train_dl), loss.item()))
        
        epoch_time = time.time() - epoch_start
        #print("[TST] Epoch: {} Train Loss: {} Time: {:.0f}:{:.0f}".format(epoch+1, epoch_loss,
        #                                                            epoch_time // 60, 
        #                                                            epoch_time % 60))
        del batch
        del inputs
        del labels
        torch.cuda.empty_cache()
    epoch_loss = (sum(losses) / len(losses))
    return epoch_loss
```

```python
def val_phase():
    model.train()
    losses = []
    epoch_start = time.time()
    for i, batch in enumerate(val_dl):
        inputs, labels = batch[0], batch[1].long()
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()             # 1. Zero the parameter gradients
        outputs = model(inputs)           # 2. Run the model

        loss = criterion(outputs, labels) # 3. Calculate loss
        #loss = criterion(outputs, torch.max(labels, 1)[1])
        losses.append(loss.item())
        #loss = torch.sqrt(loss)           #    -> RMSE loss
        loss.backward()                   # 4. Backward propagate the loss
        optimizer.step()                  # 5. Optimize the network

        #print("--> Batch {}/{} Loss: {}".format(i+1, len(train_dl), loss.item()))
        
        epoch_time = time.time() - epoch_start
        #print("[TST] Epoch: {} Train Loss: {} Time: {:.0f}:{:.0f}".format(epoch+1, epoch_loss,
        #                                                            epoch_time // 60, 
        #                                                            epoch_time % 60))
        del batch
        del inputs
        del labels
        torch.cuda.empty_cache()
    epoch_loss = (sum(losses) / len(losses))
    return epoch_loss
```

```python
def test_phase():
    model.eval()
    losses = []
    epoch_start = time.time()
    for i, batch in enumerate(unseen_dl):
        inputs, labels = batch[0], batch[1].long()
        inputs, labels = inputs.to(device), labels.to(device)
        #optimizer.zero_grad()             # 1. Zero the parameter gradients
        outputs = model(inputs)           # 2. Run the model

        loss = criterion(outputs, labels) # 3. Calculate loss
        #loss = criterion(outputs, torch.max(labels, 1)[1])
        losses.append(loss.item())
        #loss = torch.sqrt(loss)           #    -> RMSE loss
        #loss.backward()                   # 4. Backward propagate the loss
        #optimizer.step()                  # 5. Optimize the network

        #print("--> Batch {}/{} Loss: {}".format(i+1, len(train_dl), loss.item()))
        
        epoch_time = time.time() - epoch_start
        #print("[TST] Epoch: {} Train Loss: {} Time: {:.0f}:{:.0f}".format(epoch+1, epoch_loss,
        #                                                            epoch_time // 60, 
        #                                                            epoch_time % 60))
        del batch
        del inputs
        del labels
        torch.cuda.empty_cache()
    epoch_loss = (sum(losses) / len(losses))
    return epoch_loss
```

```python
model = models.resnet50(pretrained=True)
model.fc = nn.Sequential(nn.Linear(2048,1024),
                         nn.ReLU(),
                         nn.Dropout(0.9),
                         nn.Linear(1024, 512),
                         nn.ReLU(),
                         nn.Dropout(0.9),
                         nn.Linear(512,10),
                         nn.Softmax(dim=1))

model = model.to(device)
```

```python
model.fc._modules
```

```python
criterion = nn.CrossEntropyLoss()

optimizer = optim.RMSprop(model.parameters(), lr=0.001, alpha=0.99, eps=1e-08, weight_decay=5e-4, momentum=0.9, centered=False)
scheduler = lr_scheduler.ReduceLROnPlateau(optimizer, 'min', factor=0.5, verbose=True)

#RMSprop(params, lr=0.01, alpha=0.99, eps=1e-08, weight_decay=0, momentum=0, centered=False)
#Adam(model.parameters(), lr=0.9)
#SGD(model.parameters(), lr=0.001, weight_decay=5e-4, momentum=0.9)


#CyclicLR(optimizer,base_lr=0.01,max_lr=0.3,step_size_up=200)
#CosineAnnealingLR(optimizer, len(train_dl),eta_min=0.0000001)
#ReduceLROnPlateau(optimizer, 'min',factor=0.5, verbose=True)
#StepLR(optimizer, step_size=3, gamma=0.1)
```

```python
train_losses = []
val_losses= []
unseen_losses = []
for epoch in range(5):
    train_loss = train_phase()  
    #scheduler.step(train_loss)
    train_losses.append(train_loss)
    print("[TST] Epoch: {} Train Loss: {}".format(epoch+1, train_loss))
    
    val_loss = val_phase()
    scheduler.step(val_loss)       
    val_losses.append(val_loss)
    print("[TST] Epoch: {} Val Loss: {}".format(epoch+1, val_loss))
    
    unseen_loss = test_phase()      
    unseen_losses.append(unseen_loss)
    print("[TST] Epoch: {} Unseen Loss: {}".format(epoch+1, unseen_loss))
    
    #print ("LR: {}".format(scheduler.get_lr()[0]))
```

```python
plt.plot(train_losses, label='Training loss')
plt.plot(val_losses, label='Validation loss')
plt.plot(unseen_losses, label='Unseen loss')
plt.legend(frameon=False)
#plt.savefig('transf_learn_loss.jpg')
plt.show()
```

```python
model.eval()
outs = torch.Tensor().to(device)
labs = torch.Tensor().to(device)
for i, batch in enumerate(unseen_dl):
    with torch.no_grad():
        inputs, labels = batch[0], batch[1].float()
        inputs, labels = inputs.to(device), labels.to(device)
        fin_outputs = model(inputs)
        outs = torch.cat((outs,fin_outputs),0)
        labs = torch.cat((labs,labels))
```
