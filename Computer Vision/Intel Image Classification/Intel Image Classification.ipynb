```python
import os
import tensorflow as tf

from tensorflow.keras import datasets, layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from PIL import Image
```

### Let's have a look at some images

```python
all_classes = [x[0] for x in os.walk('./Data/seg_train/seg_train/')]
disp_ims = []
for class_ in all_classes[1:]:
    disp_ims.append(f'{class_}/{os.listdir(class_)[0]}')

n_row = 2
n_col = int(len(disp_ims)/2)
_, axs = plt.subplots(n_row, n_col, figsize=(12, 12))
axs = axs.flatten()
for im, class_,ax in zip(disp_ims, all_classes[1:], axs):
    title = class_.split('/')[-1]
    image = Image.open(im)
    ax.set_title(title)
    ax.imshow(image)
```

### Check image sizes

```python
for im, class_ in zip(disp_ims, all_classes[1:]):
    title = class_.split('/')[-1]
    image = Image.open(im)
    print(f'{title}: {image.size}')
```

### Let's set up our ImageDataGenerator using tensorflow and keras

```python
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory(
        './Data/seg_train/seg_train/',
        target_size=(150, 150),
        batch_size=64,
        class_mode='categorical',
        seed=42)
```

```python
validation_generator = test_datagen.flow_from_directory(
        './Data/seg_test/seg_test',
        target_size=(150, 150),
        batch_size=64,
        class_mode='categorical',
        seed=42)
```

### Let's set up the CNN 

```python
model = models.Sequential()
model.add(layers.Conv2D(16, (3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(6))
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
history = model.fit(
        train_generator,
        validation_data=validation_generator,
        epochs=10)
```

```python
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
```
