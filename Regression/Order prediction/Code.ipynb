```python
import pandas as pd
import numpy as np
import itertools
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
task1 = pd.read_csv('./archive/transact_train.txt', sep='|')
task1.head()
```

```python
task1 = task1.replace('?', np.nan)
```

```python
task1['sessionNo'].nunique()
```

```python
task1.isna().all()
```

```python
task1.dtypes
```

```python
object_to_float = ['cMinPrice', 'cMaxPrice', 'cSumPrice', 'bMinPrice', 'bMaxPrice', 'bSumPrice']
object_to_int = ['bStep', 'maxVal', 'customerScore', 'accountLifetime', 'payments', 'age', 'address', 'lastOrder', 'customerNo']
task1[object_to_float] = task1[object_to_float].astype(float)
task1[object_to_int] = task1[object_to_int].fillna(0).astype(int)
```

```python
unique_vals = task1.groupby(by=['sessionNo']).nunique().max(axis=0).reset_index()
unique_vals = unique_vals[unique_vals[0] == 1]['index'].tolist()
```

```python
unique_vals = ['sessionNo']+unique_vals
```

```python
str_cols = set(task1.select_dtypes(include=[object]).columns) - set(unique_vals)
float_cols = set(task1.select_dtypes(include=['float']).columns) - set(unique_vals)
int_cols = set(task1.select_dtypes(include=['int']).columns) - set(unique_vals)
```

```python
task1[['sessionNo']+list(str_cols)]
```

```python
task1[task1['sessionNo'] == 2][['cCount', 'cMinPrice', 'cMaxPrice', 'cSumPrice','bMinPrice','bMaxPrice', 'bSumPrice']]
```

```python
exclude = task1.groupby(by=['sessionNo'])['cCount', 'cMinPrice', 'cMaxPrice', 
                                          'cSumPrice','bMinPrice','bMaxPrice', 'bSumPrice'].sum().reset_index()
cols = exclude.iloc[:, 1:].columns.tolist()
indices = []
for col in cols:
    indices.append(exclude.iloc[:,1:][exclude.iloc[:,1:] == 0].dropna(subset=[col]).index.tolist())
merged = set(list(itertools.chain(*indices)))
ex_sesh = task1['sessionNo'].drop_duplicates().reset_index(drop=True)[merged].tolist()
new_task = task1[~task1['sessionNo'].isin(ex_sesh)].reset_index(drop=True)
```

```python
new_task.head()
```

```python
del new_task['availability']
del new_task['onlineStatus']
```

```python
float_cols
```

```python
task1.groupby(by=['sessionNo'])['customerNo'].max().tolist()
```

```python
trainingData = pd.DataFrame(columns = new_task.columns)
#trainingData['sessionNo'] = task1['sessionNo'].unique()
for col in unique_vals:
    trainingData[col] = new_task.groupby(by=['sessionNo'])[col].unique().apply(lambda x: x[0]).tolist()
for col in float_cols:
    if 'max' in col.lower():
        trainingData[col] = new_task.groupby(by=['sessionNo'])[col].max().tolist()
    if 'min' in col.lower():
        trainingData[col] = new_task.groupby(by=['sessionNo'])[col].min().tolist()
    if 'sum' in col.lower():
        trainingData[col] = new_task.groupby(by=['sessionNo'])[col].max().tolist()
    else:
        trainingData[col] = new_task.groupby(by=['sessionNo'])[col].max().tolist()
for col in int_cols:
    trainingData[col] = new_task.groupby(by=['sessionNo'])[col].max().tolist()
trainingData['order'] = trainingData['order'].map({'y': 1, 'n': 0})
```

```python
trainingData.isna().any()
```

```python
trainingData['order'].value_counts()
```

```python
trainingData.dtypes
```

```python
inputCols = trainingData.columns[1:-1].tolist()
outputCols = ['order']

X, y = trainingData[inputCols].values, trainingData[outputCols].values
X = StandardScaler().fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=42, stratify=y)
```

```python
estimator = XGBClassifier(
    objective= 'binary:logistic',
    tree_method='gpu_hist'
)
param_grid = {
    'max_depth': range (2, 10, 1),
    'n_estimators': range(60, 220, 40),
    'learning_rate': [0.1, 0.01, 0.05]
}
CV_rfc = GridSearchCV(estimator=estimator, param_grid=param_grid, cv= 5)
CV_rfc.fit(X_train, y_train)
```

```python
CV_rfc.best_estimator_
```

```python
predictions = CV_rfc.best_estimator_.predict(X_test)
classification = classification_report(y_test, predictions)
print(classification)
```

```python
def confusio_matrix(y_test, y_predicted):
  cm = confusion_matrix(y_test, y_predicted)
  plt.figure(figsize=(15,10))
  plt.clf()
  plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Wistia)
  classNames = ['Negative','Positive']
  plt.title('Confusion matrix')
  plt.ylabel('True label')
  plt.xlabel('Predicted label')
  tick_marks = np.arange(len(classNames))
  plt.xticks(tick_marks, classNames, rotation=45)
  plt.yticks(tick_marks, classNames)
  s = [['TN','FP'], ['FN', 'TP']]
  
  for i in range(2):
      for j in range(2):
          plt.text(j,i, str(s[i][j])+" = "+str(cm[i][j]))
  plt.show()
```
